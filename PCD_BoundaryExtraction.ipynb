## Import Package

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
from google.colab import files
image = files.upload()
```

## Preview Boundary Extraction

image.png

*  Erosi -> Mengurangi ukuran objek dengan mengikis tepi objek
*  Dilasi -> Memperbesar objek dengan menambahkan piksel ke tepi objek


---

Catatan:
*  Ketebalan kontur tergantung pada ukuran dari kernel.



## Inner Boundary

```python
_, binary_image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
kernel = np.ones((3,3), np.uint8)

eroded_image = cv2.erode(binary_image, kernel, iterations=1)

inner_boundary = binary_image - eroded_image

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].imshow(binary_image, cmap='gray')
axes[0].set_title('Original Image')
axes[0].axis('off')

axes[1].imshow(eroded_image, cmap='gray')
axes[1].set_title('Eroded Image')
axes[1].axis('off')

axes[2].imshow(outer_boundary, cmap='gray')
axes[2].set_title('Boundary Extraction')
axes[2].axis('off')

plt.show()
```

## Outer Boundary

```python
_, binary_image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
kernel = np.ones((3,3), np.uint8)

dilated_image = cv2.dilate(binary_image, kernel, iterations=1)

outer_boundary = dilated_image - binary_image

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].imshow(binary_image, cmap='gray')
axes[0].set_title('Original Image')
axes[0].axis('off')

axes[1].imshow(dilated_image, cmap='gray')
axes[1].set_title('Dilated Image')
axes[1].axis('off')

axes[2].imshow(inner_boundary, cmap='gray')
axes[2].set_title('Boundary Extraction')
axes[2].axis('off')

plt.show()
```
